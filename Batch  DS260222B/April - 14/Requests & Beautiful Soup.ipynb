```python
url = 'https://www.blog.pythonlibrary.org/'

def get_articles():
    
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html , 'html.parser')
    pages = soup.findAll('h2')
    
    articles = {i.a['href']: i.text.strip() for i in pages if i.a}
    
    for article in articles:
        
        s = f"{articles[article].encode('utf-8')}: {article}"
        print(s)
        
    return articles
```
