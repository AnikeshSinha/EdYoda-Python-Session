### `Check Give list is Sorted or not using Recursion`

```python
# In This Function we have make a copy of same array(list)
def isSorted(arr):
    l = len(arr)
    if l == 0 or l == 1:
        return True
    
    if arr[0] > arr[1]:
        return False
    
    
    smallerList = arr[1:]
    isSmallarlistSorted = isSorted(smallerList)
    
    #return isSmallarlistSorted
    
    if isSmallarlistSorted:
        return True
    else:
        return False 
    
#     return isSorted(arr[1:])
```

```python
arr = [1,2,3,4,5,6,7,8,9]
isSorted(arr)
```

```python
arr = [1,2,3,40,5,6,7,8,9]
isSorted(arr)
```

```python
# this solution is Far more better than Previous one Because we haven't make a copy of array each time recursion calling
# we take a advantange of list is mutable so we can use a startIndex approach to check all these element without making 
# copy of each time.

def isSortedBetter(arr , si):
    l = len(arr)
    if (si == l - 1) or (si == l):
        return True
    
    if arr[si] > arr[si + 1]:
        return False
    
    isSmallarPartSorted = isSortedBetter(arr ,si + 1)
    return isSmallarPartSorted
```

```python
arr = [1,2,3,40,5,6,7,8,9]
isSortedBetter(arr , 0)
```

```python
arr = [1,2,3,4,5,6,7,8,9]
isSortedBetter(arr , 0)
```

### `Replace Character a by character b in given String`

```python
def replaceChar(s , a , b ): # here a , b is a parameter and s is a string
    if len(s) == 0:
        return s
    
    smallOutput = replaceChar(s[1:] , a , b)
    if s[0] == a:
        return b + smallOutput

    else:
        return s[0] + smallOutput
```

```python
replaceChar("cc" , "a" , "b")
```

```python
replaceChar("Tejashree" , "e" , "s")
```

```python
replaceChar("x" , "c" , "x")
```

```python
replaceChar("xcx" , "c" , "x")
```

```python
replaceChar("cccc" , "c" , "x")
```

### `Replace Character pi with 3.14 in Given String`

```python
def replacePi(s):
    if len(s) == 0 or len(s) == 1:
        return s
    
    if s[0] == 'p' and s[1] == 'i':
        smallOutput = replacePi(s[2:])  # here if both p and i available in the begining then we can call our
                                            #  recursive fuction after 3rd Element of the string
        return "3.14" + smallOutput    # here We just Concatinate With smallOutput we get from recursion calling.
    
    else:
        smallOutput = replacePi(s[1:])  # here if p and i is not the First 2 character in given string then it we 
                                        # have to call our recursive function with 1st element of the given string
        return s[0] + smallOutput  # here we simply conctinate with First Character with SmallOutput.
    
```

```python
replacePi("ppipipi")
```

```python
replacePi("abcd")
```

```python
replacePi("pabcd")
```

```python
replacePi("pi")
```

```python
replacePi("pippi")
```
